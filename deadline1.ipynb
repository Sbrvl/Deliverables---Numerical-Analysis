# Bibliographic Research:

## (a) In which contexts does the KPP equation arise?

The Kolmogorov-Petrovsky-Piskunov (KPP) equation, also known as the Fisher-KPP equation, appears in various scientific and engineering contexts, including:

- **Population Genetics and Ecology**: Originally introduced to describe gene frequency propagation in spatially distributed populations, capturing how advantageous genes spread through a habitat. ([Fisher, 1937](https://doi.org/10.1111/j.1469-1809.1937.tb02153.x)).
- **Reaction-Diffusion Systems**: Used in modeling chemical and biological fronts where substances or species diffuse and react, leading to traveling wave solutions. ([Embree, 2001](https://people.maths.ox.ac.uk/trefethen/pdectb/fisher2.pdf)).
- **Combustion Theory**: Appears in the study of flame propagation in premixed gases, where heat and reactants diffuse while undergoing exothermic reactions. ([Zeldovich & Frank-Kamenetskii, 1938](https://doi.org/10.1088/0959-5309/53/4/301)).
- **Neuroscience**: Models the propagation of action potentials in nerve fibers, describing the dynamics of neural excitation and signal transmission. ([Corti et al., 2023](https://www.mate.polimi.it/biblioteca/add/qmox/41-2023.pdf)).
- **Epidemiology**: Describes how infectious diseases spread spatially in a population, illustrating how diseases can propagate over geographical areas. ([Murray, 2002](https://doi.org/10.1007/978-3-662-08539-4)).

## (b) How is the KPP equation derived from these contexts?

The KPP equation is a reaction-diffusion equation given by:

$$
\frac{\partial u}{\partial t} = \alpha F(u) + \beta \frac{\partial^2 u}{\partial x^2} + f
$$

where:
- $u(t, x)$ represents the quantity of interest (e.g., population density, concentration, etc.).
- $\alpha F(u)$ represents the reaction term.
- $\beta \frac{\partial^2 u}{\partial x^2}$ models diffusion.
- $f$ represents external forcing.

The derivation of the KPP equation depends on the specific application:

- **In Population Genetics**: The equation arises from a **diffusion model of gene frequencies**, originally formulated by Fisher and later extended by Kolmogorov, Petrovskii, and Piskunov.

- **In Chemical and Biological Reaction-Diffusion Systems**: The equation is derived using **Fickâ€™s Law of Diffusion** and reaction kinetics, leading to a **nonlinear PDE** describing how a substance spreads while reacting.

- **In Combustion Theory**: The derivation involves **heat conduction equations coupled with chemical reaction rates**, where the reaction-diffusion form naturally emerges in **flame propagation**.

- **In Neuroscience**: The equation models **wave propagation of neural action potentials**, and it can be derived from **Hodgkin-Huxley-type models** under certain simplifications.

- **In Epidemiology**: The KPP equation emerges in **disease spread models**, particularly in spatially dependent **SIR (Susceptible-Infected-Recovered) models** with diffusion.

For a general mathematical treatment of the KPP equation and its derivation in different physical and biological contexts, see **[Aronson & Weinberger, 1975](https://doi.org/10.2307/1997287)**.

## (c) What are the numerical methods of resolution of the KPP equation?

Common numerical methods used to solve the KPP equation include:

- **Finite Difference Methods (FDM)**:
  - Explicit and implicit schemes for time discretization.
  - Crank-Nicholson method for stability ([Ibrahim, 2020](https://lutpub.lut.fi/bitstream/handle/10024/161384/msc_thesis_ibrahim.pdf?sequence=1)).
  
  
- **Finite Element Methods (FEM)**:
  - Useful for irregular geometries ([Ern & Guermond, 2004](https://doi.org/10.1007/978-1-4757-4355-5)).
  
- **Spectral Methods**:
  - High-accuracy solutions for periodic boundary conditions ([Singh et al., 2022](https://dr.ntu.edu.sg/bitstream/10356/156096/2/Chapter11_Satyvir%20Singh.pdf)).

- **Monte Carlo Methods**:
  - Used in stochastic versions of the equation ([Garnier & Souganidis, 1998](https://doi.org/10.1007/s002850050091)).


# Questions on the KPP Equation: 

## (a) Is the KPP equation an ODE or a PDE? Is it linear, homogeneous, autonomous?

The KPP equation is a **partial differential equation (PDE)** since it involves derivatives with respect to both **time** and **space**:

$$
\frac{\partial u}{\partial t} = \alpha F(u) + \beta \frac{\partial^2 u}{\partial x^2} + f.
$$

Given the **nonlinear term**:

$$
F(u) = u \frac{u_0 - u}{u_0 + u}
$$

and the **source term**:

$$
f(t,x) = T_0 \tanh\left(\gamma \frac{x}{L}\right) \frac{2}{\pi} \arctan\left(\frac{t}{\tau}\right), \quad \text{for } \gamma > 0,
$$

we analyze the properties:

- **Nonlinear**: The reaction term $F(u)$ is **nonlinear** due to the fraction involving $u$.
- **Inhomogeneous**: The presence of the source term $f(t, x)$ makes it **inhomogeneous**.
- **Not Autonomous**: The equation **is not strictly autonomous** due to the explicit dependence of $f(t, x)$ on $t$.

## (b) What is the physical dimension (or unit) of $\alpha$, $\beta$, $\sqrt{\alpha \beta}$, and $\tau$?

- $u$ is typically **dimensionless** (e.g., a normalized population density).
- $x$ has **units of length** ($L$).
- $t$ has **units of time** ($T$).

From the equation:

$$
\frac{\partial u}{\partial t} = \alpha F(u) + \beta \frac{\partial^2 u}{\partial x^2} + f,
$$

we determine the physical dimensions:

- $\alpha$ has units of **inverse time**: $[\alpha] = T^{-1}$.
- $\beta$ has units of **diffusivity**: $[\beta] = L^2 T^{-1}$.
- $\sqrt{\alpha \beta}$ has units of **velocity**:

  $$
  \sqrt{\alpha \beta} = (T^{-1} \cdot L^2 T^{-1})^{1/2} = L T^{-1}.
  $$

- $\tau$ is a **characteristic time scale**, appearing in the source term, with units of **time**: $[\tau] = T$.

## (c) If the source $f = 0$, what are the constant solutions to the equation? Which equilibria are stable/unstable?

When the source term is **absent** ($f = 0$), we solve the **steady-state equation** by setting:

$$
\alpha F(u) + \beta \frac{\partial^2 u}{\partial x^2} = 0.
$$

For spatially **constant solutions** ($\partial_x^2 u = 0$), we get:

$$
\alpha F(u) = 0.
$$

Substituting the **nonlinear term**:

$$
\alpha u \frac{u_0 - u}{u_0 + u} = 0.
$$

This equation has two solutions:

$$
u = 0, \quad u = u_0.
$$


## Stability Analysis

To determine stability, we linearize around each equilibrium solution by considering a **small perturbation** $v = u - u^*$, where $u^*$ is the equilibrium. First, we derive the expression for $F'(u)$ for:

$$
F(u) = \frac{u\,(u_0 - u)}{u_0 + u}.
$$

### **Derivation of \(F'(u)\)**

Let

$$
g(u) = u\,(u_0 - u), \quad h(u) = u_0 + u.
$$

Then, by the quotient rule:

$$
F'(u) = \frac{g'(u)h(u) - g(u)h'(u)}{[h(u)]^2},
$$

with

$$
g'(u) = u_0 - 2u, \quad h'(u) = 1.
$$

Thus,

$$
F'(u) = \frac{(u_0 - 2u)(u_0 + u) - u(u_0 - u)}{(u_0 + u)^2}.
$$

### **1. For \(u = 0\):**

- Small perturbation: $u = 0 + v$.
- Linearizing $F(u)$ around $u=0$ gives:

  $$
  F(u) \approx F(0) + F'(0)v.
  $$

- At $u=0$:

  $$
  F(0) = \frac{0\,(u_0 - 0)}{u_0} = 0,
  $$

  and

  $$
  F'(0) = \frac{(u_0)(u_0) - 0}{(u_0)^2} = 1.
  $$

- Therefore, near $u=0$:

  $$
  F(u) \approx v,
  $$

  and the linearized equation is

  $$
  \frac{dv}{dt} = \alpha\,F'(0)v = \alpha\,v.
  $$

  Since $\alpha > 0$, the perturbation $v$ grows exponentially, indicating that **$u = 0$ is unstable**.



### **2. For \(u = u_0\):**

- Small perturbation: $u = u_0 + v$.
- Linearizing $F(u)$ around $u=u_0$ gives:

  $$
  F(u) \approx F(u_0) + F'(u_0)v.
  $$

- At $u=u_0$:

  $$
  F(u_0) = \frac{u_0\,(u_0 - u_0)}{u_0+u_0} = 0,
  $$

  and

  $$
  F'(u_0) = \frac{(u_0 - 2u_0)(u_0 + u_0) - 0}{(2u_0)^2} = \frac{(-u_0)(2u_0)}{4u_0^2} = -\frac{1}{2}.
  $$

- Therefore, near $u=u_0$:

  $$
  F(u) \approx -\frac{1}{2}\,v,
  $$

  and the linearized equation is

  $$
  \frac{dv}{dt} = \alpha\,F'(u_0)v = -\frac{\alpha}{2}\,v.
  $$

  Since $-\frac{\alpha}{2} < 0$, the perturbation $v$ decays exponentially, indicating that **$u = u_0$ is stable**.
