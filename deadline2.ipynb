## 1) Show that the IVP has a unique solution $y \in C^1([0,T],E)$


Since $f:[0,T]\times E \to E$ is continuous and $L$-Lipschitz with respect to its second variable, the Cauchy–Lipschitz theorem guarantees that the system:

$$
\begin{cases}
y'(t) = f\bigl(t,y(t)\bigr), \\[6pt]
y(0) = y^0
\end{cases}
$$

has a unique solution $y$ in $E$.



## 2) Show that for $T=1,\ f:(t,y)\in [0,1]\times E \mapsto 2y \in E,\ \text{and } N=0,\ z^1 \text{ is not well-defined}$

For $T=1$ and $N=0$, $h = \dfrac{1}{0+1} = 1.$

For all $k \in \{0,1\}$, we have $t^k = k \, h.$ Thus $t^0 = 0$ and $t^1 = 1.$

We set $z^0 = y^0.$

Then,
$$
z^1 \;=\; z^0 \;+\; \frac{h}{2} \,\bigl[\,2\,z^0 \;+\; 2\,z^1\bigr] \;=\; z^0 \;+\; \frac{1}{2} \,\bigl[\,2\,z^0 \;+\; 2\,z^1\bigr].
$$

This implies:
$$
z^1 
\;=\; z^0 + z^0 + z^1 
\;\;\Longrightarrow\;\; 2\,z^0 = 0 
\;\;\Longrightarrow\;\; z^0 = 0 
\;\;\Longrightarrow\;\; z^0 = y^0 = 0.
$$

Hence, if $y^0 \neq 0,$ then $z^1$ is not defined. Therefore, $z^1$ is only defined if $z^0 = 0,$ i.e., $y^0 = 0.$



## 3) Show that if $hL < 2$, then the Crank–Nicolson scheme is well-defined

$z^{0} = y^{0}$ is well-defined.

Let $n \in \{0,\dots,N-1\}$ and suppose that $z^0,\dots,z^n$ are well-defined. 

Let us show that $z^{n+1}$ is well-defined. This is true if and only if there exists a unique solution to:

$$
z^{n+1} = z^n + \frac{h}{2}\Bigl[f\bigl(t^n, z^n\bigr) + f\bigl(t^{n+1}, z^{n+1}\bigr)\Bigr].
$$

We introduce

$$
g_n : x \in E \longrightarrow E,\quad g_n(x) = z^n + \frac{h}{2}\Bigl[f\bigl(t^n, z^n\bigr) + f\bigl(t^{n+1}, x\bigr)\Bigr].
$$

We have that $z^{n+1}$ is well-defined if and only if there exists a unique $x \in E$ such that

$$
g_n(x) = x,
$$

i.e., there is a unique fixed point for $g_n$.

By Banach's Fixed Point Theorem, there is a unique fixed point for $g_n$ if there exists $\alpha < 1$ such that $g_n$ is $\alpha$-Lipschitz. 

Let $x_1,x_2 \in E$. Then,

$$
\|g_n(x_1) - g_n(x_2)\| = \left\| \left(z^n + \frac{h}{2}\Bigl[f(t^n,z^n) + f(t^{n+1},x_1)\Bigr]\right) - \left(z^n + \frac{h}{2}\Bigl[f(t^n,z^n) + f(t^{n+1},x_2)\Bigr]\right) \right\|
$$

$$
\quad = \frac{h}{2}\,\|f(t^{n+1},x_1) - f(t^{n+1},x_2)\|.
$$

Since $f$ is $L$-Lipschitz with respect to its second variable, we have

$$
\|g_n(x_1) - g_n(x_2)\| \le \frac{hL}{2}\,\|x_1 - x_2\|.
$$

If $hL < 2$, then $\frac{hL}{2} < 1$, so $g_n$ satisfies the contraction condition of Banach's Fixed Point Theorem. Hence, there is a unique fixed point, which defines $z^{n+1}$ uniquely.

Therefore, the **Crank–Nicolson scheme** is well-defined.

## 4) Show that the Crank–Nicolson scheme is consistent with the IVP for the 1-norm


Using Taylor's theorem, we have for some $\theta^n \in [t^{n-1}, t^n]$
$$
y(t^{n}) = y(t^{n-1} + h) = y(t^{n-1}) + hy'(t^{n-1}) + \frac{h^2}{2}y''(t^{n-1}) + \frac{h^3}{6}\,y'''(\theta^n).
$$
We have the consistency error $e^{n}$:
$$
e^{n} = y(t^{n}) - y(t^{n-1}) + \frac{h}{2}\Bigl[f(t^{n-1}, y(t^{n-1})) - f(t^n,y(t^n))\Bigr] = y(t^{n}) - y(t^{n-1}) + \frac{h}{2}\Bigl[y'(t^{n-1}) + y'(t^{n})\Bigr].
$$
Again, using Taylor's theorem, we have for some $\xi^n \in [t^{n-1}, t^n]$:
$$
y'(t^{n}) = y'(t^{n-1} + h) = y'(t^{n-1}) + hy''(t^{n-1}) + \frac{h^2}{2}\,y'''(\xi^n).
$$
We insert $y'(t^{n})$ and $y(t^{n})$ expanded forms into the consistency error $e^{n}$ and get:
$$
e^{n} = hy'(t^{n-1}) + \frac{h^2}{2}y''(t^{n-1}) + \frac{h^3}{6}\,y'''(\theta^n) - \frac{h}{2}\Bigl[2y'(t^{n-1}) + hy''(t^{n-1}) + \frac{h^2}{2}\,y'''(\xi^n)\Bigr] = \frac{h^3}{6}\,y'''(\theta^n) -  \frac{h^3}{4}\,y'''(\xi^n).
$$
Taking the absolute value and using the triangle inequality we get:
$$
\Bigl|\frac{h^3}{6}\,y'''(\theta^n) -  \frac{h^3}{4}\,y'''(\xi^n)\Bigr| \leq h^3 \Bigl( \Bigl|\frac{y'''(\theta^n)}{6} \Bigr|  +  \Bigl|\frac{y'''(\xi^n)}{4} \Bigr| \Bigr)
$$
Denote $M = \sup_{1 \leq n \leq N+1} \Bigl( \Bigl|\frac{y'''(\theta^n)}{6} \Bigr|  +  \Bigl|\frac{y'''(\xi^n)}{4} \Bigr| \Bigr)$ . Now, taking the 1-norm and using the L-Lipschitz property of f, it follows that:
$$
\| e_{n+1} \|_1
\;\;\le\;\;
\sum_{i=1}^{N+1} h^3 \Bigl|\frac{y'''(\theta^n)}{6}\ -  \frac{y'''(\xi^n)}{4} \Bigr|
\;\;\le\;\;
h^3 \sum_{i=1}^{N+1} M
\;=\;
h^3\,M\,(N+1)
\;=\;
\Bigl(\frac{T}{N+1}\Bigr)^3\,M\,(N+1)
\;=\;
\frac{T^3\,M}{(N+1)^2}
\;\;\longrightarrow\;\;
0
\quad\text{as}\quad
N\to\infty.
$$
Thus the consistency error tend to 0 as N tend to infinity.

## 7) Implement in Python a function `crank_nicolson` in $\mathbb{R}^d$

```python
import numpy as np

def crank_nicolson(f, t0, T, y0, N):
    """
    Crank–Nicolson solver for y'(t) = f(t, y(t)) in R^d.
    
    Parameters:
    -----------
    f   : callable, f(t, y) -> array_like
    t0  : float, initial time
    T   : float, final time
    y0  : array_like, initial condition in R^d
    N   : int, number of time steps

    Returns:
    --------
    t   : numpy array of shape (N+1,) with the time grid
    Y   : numpy array of shape (N+1, d) with the solution at each time
    """
    d = len(y0)
    h = (T - t0) / N
    t = np.linspace(t0, T, N+1)
    
    Y = np.zeros((N+1, d))
    Y[0] = y0
    
    for n in range(N):
        tn = t[n]
        y_pred = Y[n] + h * f(tn, Y[n])  # explicit Euler predictor
        x_next = y_pred.copy()
        for _ in range(5):  # simple fixed-point iteration
            x_next = Y[n] + 0.5 * h * ( f(tn, Y[n]) + f(tn + h, x_next) )
        Y[n+1] = x_next
    
    return t, Y
